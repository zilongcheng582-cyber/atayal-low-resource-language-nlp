# 02 Tokenization Pipeline — Atayal (Zeyuan Dialect)
This notebook converts the cleaned Atayal text into numerical sequences for LSTM training.
## Tokenization 结果摘要
- 总字符数：58,437
- 字符集大小：61
- SEQ_LEN：50
- 训练样本数：58,387

```python
import urllib.request

for i in range(1, 13):
    url = f"https://raw.githubusercontent.com/zilongcheng582-cyber/atayal-nlp-summer/main/data/raw/{i}.pyu"
    urllib.request.urlretrieve(url, f"{i}.pyu")

print("下载完成！")
```

```python
import string

# 读取数据
all_lines = []
for i in range(1, 13):
    filename = f"{i}.pyu"
    with open(filename, "r", encoding="utf-8") as f:
        lines = f.readlines()
    lines = [line.strip() for line in lines if line.strip()]
    all_lines.extend(lines)

# 清洗
def has_chinese(text):
    return any('\u4e00' <= char <= '\u9fff' for char in text)

clean_lines = [line for line in all_lines if not has_chinese(line)]

# 字符集 注意：这里用 \n 连接以保留句子边界，与 corpus analysis 中不同
full_text = "\n".join(clean_lines)
chars = sorted(set(full_text))
char2idx = {char: idx for idx, char in enumerate(chars)}
idx2char = {idx: char for idx, char in enumerate(chars)}

print(f"句子数：{len(clean_lines)}")
print(f"字符集大小：{len(char2idx)}")
```

```python
# 用换行符连接所有句子
full_text = "\n".join(clean_lines)
print(f"总字符数：{len(full_text)}")
print(f"前50个字符：{full_text[:50]}")
```

```python
# 把整个文本编码成数字列表
encoded = [char2idx[c] for c in full_text]

print(f"编码后长度：{len(encoded)}")
print(f"前10个数字：{encoded[:10]}")
print(f"对应字符：{[idx2char[i] for i in encoded[:10]]}")
```

```python
SEQ_LEN = 50  # 每个输入序列的长度

inputs = []
targets = []

for i in range(0, len(encoded) - SEQ_LEN):
    inputs.append(encoded[i : i + SEQ_LEN])
    targets.append(encoded[i + 1 : i + SEQ_LEN + 1])

print(f"训练样本数：{len(inputs)}")
print(f"第一个input：{inputs[0]}")
print(f"第一个target：{targets[0]}")
```

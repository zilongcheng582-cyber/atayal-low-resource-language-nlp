# 01 Corpus Analysis — Atayal (Zeyuan Dialect)
This notebook performs basic statistical analysis on the Atayal text corpus.
## 分析结果摘要 (All Chapters, N=12)
- 原始句子数：689
- 清洗后句子数：688（过滤1句含中文噪声）
- 总词数：10,566
- 词汇表大小：1,846
- 平均句子长度：15.4 个词（最短1词，最长161词）
- 字符集大小：60
- 最高频词：ga(432), ku(368), na(290), ru(264), ta(177)


```python
# 读取泰雅语语料
all_lines = []

for i in range(1, 13):
    filename = f"{i}.pyu"
    with open(filename, "r", encoding="utf-8") as f:
        lines = f.readlines()
    lines = [line.strip() for line in lines if line.strip()]
    all_lines.extend(lines)

print(f"总句子数：{len(all_lines)}")
print("前3句：")
for line in all_lines[:3]:
    print(line)
```

```python
# 过滤掉含中文的句子
def has_chinese(text):
    return any('\u4e00' <= char <= '\u9fff' for char in text)

clean_lines = [line for line in all_lines if not has_chinese(line)]

print(f"清洗前句子数：{len(all_lines)}")
print(f"清洗后句子数：{len(clean_lines)}")
```

```python
import string

sentence = "psba biru, lawkah su ga?"

# string.punctuation 包含所有标点
# 但我们要把 ' 从里面移除，因为泰雅语需要它
punctuation_to_remove = string.punctuation.replace("'", "")

# 去掉标点
cleaned = sentence.translate(str.maketrans("", "", punctuation_to_remove))

print(cleaned)
```

```python
import string
from collections import Counter

# 去掉标点（保留撇号）
punctuation_to_remove = string.punctuation.replace("'", "")

# 处理所有句子，分割成词
all_words = []
for line in clean_lines:
    cleaned = line.translate(str.maketrans("", "", punctuation_to_remove))
    words = cleaned.lower().split()
    all_words.extend(words)

# 统计
word_counts = Counter(all_words)

print(f"总词数（含重复）：{len(all_words)}")
print(f"词汇表大小（不重复）：{len(word_counts)}")
print(f"\n最常见的10个词：")
for word, count in word_counts.most_common(10):
    print(f"  {word}: {count}")
```

```python
# 每句话的词数
sentence_lengths = [len(line.translate(
    str.maketrans("", "", punctuation_to_remove)).lower().split())
    for line in clean_lines]

avg_length = sum(sentence_lengths) / len(sentence_lengths)
print(f"平均句子长度：{avg_length:.1f} 个词")
print(f"最短句子：{min(sentence_lengths)} 个词")
print(f"最长句子：{max(sentence_lengths)} 个词")
```

```python
# 把所有文本合并成一个大字符串
full_text = " ".join(clean_lines)

# 找出所有不重复的字符
chars = sorted(set(full_text))

print(f"字符集大小：{len(chars)}")
print(f"所有字符：{''.join(chars)}")
```

```python
char2idx = {char: idx for idx, char in enumerate(chars)}
idx2char = {idx: char for idx, char in enumerate(chars)}

print(f"字符集大小：{len(char2idx)}")
print(f"前5个映射：{ {k: char2idx[k] for k in list(char2idx)[:5]} }")
```
